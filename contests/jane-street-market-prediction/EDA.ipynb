```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%%time
df_head = pd.read_csv('train_head.csv', index_col='ts_id')
df_head.columns.values
```

```python
%%time
dataset = pd.read_csv('train.csv', index_col='ts_id')
```

```python
%%time
df = dataset.copy()
## drop NaN-containing columns for now
df.drop(labels=df.columns[df.isna().any()].tolist(), axis=1, inplace=True)
```

```python
%%time
anon_features = [col for col in df.columns if 'feature' in col]
corr = df[anon_features].corr()
```

```python
f, ax = plt.subplots(figsize=(21, 15))

cmap = sns.diverging_palette(20, 230, as_cmap=True)
sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
