```python
%%time
sales = sales_original.copy()
sales['time'] = pd.to_datetime(sales.date, format='%d.%m.%Y')
sales.drop('date', inplace = True, axis = 1)
sales['month'] = sales.time.dt.month
sales['week'] = sales.time.dt.week
sales['weekday'] = sales.time.dt.weekday
sales = sales.set_index('time')
```
